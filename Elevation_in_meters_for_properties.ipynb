```python
import pandas as pd
import numpy as np
import requests

key = 'INSERT_GOOGLE_MAPS_API_KEY'
```

```python
df = pd.read_csv('data/trump_real_estate_portfolio.csv')
```

```python
df
```

```python
def create_elevation_url(row):
    return 'https://maps.googleapis.com/maps/api/elevation/json?locations=' + str(row['Lat']) + ',' + str(row['Long']) + '&key='
```

```python
df['elevation_url'] = df.apply(lambda row: create_elevation_url(row),axis=1)
```

```python
def get_elevation(row, key):
    r = requests.get(row['elevation_url'] + key)
    try:
        return r.json()['results'][0]['elevation']
    except IndexError:
        return r.json()['results']
```

```python
df['elevation_in_meters'] = df.apply(lambda row: get_elevation(row, key),axis=1)
```
